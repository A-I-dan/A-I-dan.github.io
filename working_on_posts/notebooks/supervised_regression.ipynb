```python
# Regression Example 
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

X, y = make_regression(n_samples=100, n_features=1, noise=13)

plt.scatter(X,y)
plt.show()
```

```python
reg = LinearRegression(fit_intercept=True)
reg.fit(X, y)

xfit = np.linspace(start=-3, stop =3)
yfit = reg.predict(xfit[:, np.newaxis])

plt.scatter(X, y)
plt.plot(xfit, yfit)
```

```python
print("Slope:    ", reg.coef_[0])
print("Intercept:", reg.intercept_)
```

```python
def predict(x):
    w = reg.coef_[0]
    b = reg.intercept_
    y = w*x + b
    return y
```

```python
print("Predict y For 3:     ", predict(3))
print("Predict y For -1.5: ", predict(-1.5))
```

```python
plt.scatter(X, y)
plt.plot(xfit, yfit)
plt.plot(3, predict(3), 'ro')
plt.plot(-1.5, predict(-1.5), 'ro')
```
